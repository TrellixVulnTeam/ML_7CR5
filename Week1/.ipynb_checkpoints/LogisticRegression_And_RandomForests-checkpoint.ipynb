```python
#Importing the basic needed libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Read data for Linear Regression from csv

df=pd.read_csv("LoR_DT_RF\\train.csv")
```

```python
df
```

# Handling Null values

```python
# Trying to get an approx. idea how many columns have a major amount of missing data

fig, ax = plt.subplots(figsize=(20,8))
sns.heatmap(df.isnull(),cmap='coolwarm')
```

```python
df.groupby('Pclass')['Age'].mean()
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 38

        elif Pclass == 2:
            return 30

        else:
            return 25

    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
df['Embarked'].isnull().sum()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked']=df['Embarked'].fillna('S')
```

```python
df.info()
```

```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
df['Ticket'].value_counts()
```

```python
df.drop(['Ticket'],axis=1,inplace=True)
```

```python
new_cols={}
drop_cols=[]
for col in df.columns:
    if(df.dtypes[col]=='O'):
        drop_cols.append(col)
        new_cols[col+'_cols'] = pd.get_dummies(df[col], prefix=col, drop_first=True)
for col in new_cols.keys():        
    df = pd.concat([df,new_cols[col]],axis=1)
df.drop(drop_cols,axis=1,inplace=True)
```

```python
df
```

```python
#Using corr matrix
fig, ax = plt.subplots(figsize=(7,6))
sns.heatmap(df.corr(),linecolor='white',linewidths=1,cmap='magma',annot=True)
```

```python
#import Logistic Regression

from sklearn.model_selection import train_test_split

X=df.drop('Survived',axis=1)
y=df['Survived']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)

predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```

# RANDOM FOREST

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth':np.linspace(1, 32, 32),'min_samples_split':np.linspace(0.1, 1.0, 10),'n_estimators':np.arange(1,1000,10) }
clf = GridSearchCV(RandomForestClassifier(), param_grid,refit=True)
clf.fit(X_train,y_train)
```
